```python
import pandas as pd
import threading
import time

# Load dataset
df = pd.read_csv("C:/Users/hp/Downloads/Delhi metro.csv")
df["Station Names"] = df["Station Names"].str.strip().str.lower()

# Group by metro line
line_map = df.groupby("Metro Line")["Station Names"].apply(list).to_dict()

# Mutex for shared station
mutex = threading.Lock()

# Simulate train running on a line
def run_train(line_name, train_id):
    stations = line_map[line_name]
    for station in stations:
        with mutex:  # Critical section
            print(f"[Train {train_id}] arriving at {station.title()} on {line_name}")
        time.sleep(1)  # simulate travel delay
    print(f"[Train {train_id}] finished line {line_name}")

# Simulate two or more trains (threads)
def main():
    print("Simulating Trains as Threads...")
    threads = []

    # Choose 2 metro lines for simulation
    selected_lines = list(line_map.keys())[:2]
    for i, line in enumerate(selected_lines):
        t = threading.Thread(target=run_train, args=(line, i+1))
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    print("Simulation complete.")

if __name__ == "__main__":
    main()
```
